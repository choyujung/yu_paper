```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
play_id = []
play_name = []
# play_time = []
# age_limit = []
# play_price = []
# stage_type = []
# seat_scale = []
# play_cnt = []
# image_url_list = []
play_df = ""
```

```python
apikey = 'd2d444505af640df8e926c51e32aebca'
stdate = '20160201'
eddate = '20161231'
url = 'http://www.kopis.or.kr/openApi/restful/prfstsPrfBy?service='+apikey+'&cpage=1&rows=399&stdate='+stdate+'&eddate='+eddate+'&shcate=AAAA'
```

```python
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# #공연횟수
# count_list = soup.select('prfdtcnt')
# for count in count_list:
#     play_cnt.append(count.text)
```

```python
id_list = soup.select('mt20id')
for i in id_list:
    play_id.append(i.text)
temp = {'연극ID':play_id}
play_df = pd.DataFrame(temp)
```

```python
items = soup.select('prfnm')
for item in items:
    play_name.append(item.text)
play_df['연극제목'] = play_name
play_df
```

```python
stage_info_list = []
```

```python
age_limit = []
play_price = []
image_url_list = []
play_time = []
stage_info_list = []
openrun_list = []
daycnt_list = []
actor_list = []
```

```python
#play_df=play_df.drop(['openrun','요일수','배우'],axis=1)
```

```python
for name in play_id:
    detail_url = 'http://www.kopis.or.kr/openApi/restful/pblprfr/'+name+'?service='+apikey
    detail_response = requests.get(detail_url)
    detail_soup = BeautifulSoup(detail_response.text, 'html.parser')
    age_list = detail_soup.select('prfage')
    age_limit.append(age_list[0].text)
    price_list = detail_soup.select('pcseguidance')
    play_price.append(price_list[0].text)
    image_list = detail_soup.select('styurl')
    image_url_list.append(image_list)
    time_list = detail_soup.select('prfruntime')
    play_time.append(time_list[0].text)
    stage_info = detail_soup.select('mt10id')
    stage_info_list.append(stage_info[0].text)
    
    openrun = detail_soup.select('openrun')
    if len(openrun) == 0:
        openrun_list.append(0)
    else :
        openrun_list.append(openrun[0].text)
    daycnt = detail_soup.select('dtguidance')
    daycnt_list.append(daycnt[0].text)
    actor = detail_soup.select('prfcast')
    actor_list.append(actor[0].text)
    
    
play_df['공연시간'] = play_time
play_df['관객나이'] = age_limit
play_df['가격'] = play_price
play_df['openrun'] = openrun_list
play_df['요일수'] = daycnt_list
play_df['배우'] = actor_list


play_df
```

```python
# stage_type_list = []
# seat_scale_list = []
```

```python
stagecnt_list = []
stageyear_list = []
region_list = []
```

```python
for stage in stage_info_list:
    stage_url = 'http://www.kopis.or.kr/openApi/restful/prfplc/'+stage+'?service='+apikey
    stage_response = requests.get(stage_url)
    stage_soup = BeautifulSoup(stage_response.text, 'html.parser')
    
#     stage_type = stage_soup.select('fcltychartr')
#     if len(stage_type) != 0:
#         stage_type_list.append(stage_type[0].text)
#     elif len(stage_type) == 0:
#         stage_type_list.append(0)
#     seat_scale = stage_soup.select('seatscale')
#     if len(seat_scale) != 0:
#         seat_scale_list.append(seat_scale[0].text)
#     elif len(seat_scale) == 0:
#         seat_scale_list.append(0)
        
    stagecnt = stage_soup.select('mt13cnt')
    if len(stagecnt) == 0 :
        stagecnt_list.append(0)
    else :
        stagecnt_list.append(stagecnt[0].text)
        
    stageyear = stage_soup.select('opende') 
    if len(stageyear) == 0 :
        stageyear_list.append(0)
    else:
        stageyear_list.append(stageyear[0].text)
        
    region = stage_soup.select('adres')
    if len(region) == 0 :
        region_list.append(0)
    else:
        region_list.append(region[0].text)
        
# play_df['공연장유형'] = stage_type_list
# play_df['객석수'] = seat_scale_list
# play_df['공연수'] = play_cnt
play_df['공연장수'] = stagecnt_list
play_df['공연장연혁'] = stageyear_list
play_df['수도권여부'] = region_list
```
