```python
##시작날짜만 리스트로 만들기
start_list = []
for m in range(0, 12):
    start = datetime(2021,1,1) + relativedelta(months=m)
    start_list.append(start)
```

```python
## 각 기업별로 한달씩 끊어서 rsi, roc, voliatility

no_label = []

label_df = pd.DataFrame(columns=['Date','Company','Code','Image_name', 'RSI', 'VOL'])     #'Company'

for j in tqdm(range(len(excel_list))):
  df = pd.read_excel(path + '/' + excel_list[j])
  print(excel_list[j])

  COMPANY = excel_list[j].split('_')[0]
  CODE = excel_list[j].split('_')[1].split('.')[0]
  # CODE = excel_list[j].split('.')[0]

  for i in tqdm(range(len(start_list))):
    start = start_list[i]
    end = start + relativedelta(months=1) - timedelta(days=1)
    # print(start)
    # print(end)

    IMAGE_NAME = excel_list[j].split('.')[0] + '_' + str(start_list[i].month) + '.jpg'

    df2 = df.loc[(df['Date']>=start)& (df['Date']<=end)]
    # print(len(df2))

    if (df2.empty == False) :

      if (df2['Open']==0).any() ==False :
        # rsi 계산
        try : 
          rsi = ta.momentum.rsi(df2['Close'], window = len(df2))   # default window = 14   ##추세 상승,횡보,하락
          rsi.reset_index(drop = True, inplace=True)
          RSI = rsi[len(df2)-1]
          # df['roc'] = ta.momentum.roc(df['Close'], window = len(df)-1)   # default window = 14

          VOL =  np.sqrt(df2['Close'].pct_change().var())    #변동성 4구간

        except ZeroDivisionError:
          pass

        label_df = label_df.append(pd.DataFrame([[start,COMPANY,CODE,IMAGE_NAME, RSI, VOL]], columns=['Date','Company','Code','Image_name', 'RSI', 'VOL']), ignore_index = True)    #'Company'
        
      else :
        no_label.append(excel_list[j].split('.')[0] + '_' + str(start_list[i].month))
        pass

    else : 
      no_label.append(excel_list[j].split('.')[0] + '_' + str(start_list[i].month))
      pass


  if j % 500 == 0 :
    label_df.to_excel('/content/drive/MyDrive/stock_hkh/kospi_label_df.xlsx',index = False)
    
    with open("/content/drive/MyDrive/stock_hkh/kospi_no_label.csv", 'w') as file: 
      writer = csv.writer(file) 
      writer.writerow(no_label)
```
