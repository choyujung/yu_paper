```python
##시작날짜만 리스트로 만들기
start_list = []
for m in range(0, 12):
    start = datetime(2021,1,1) + relativedelta(months=m)
    start_list.append(start)
```

```python
## 각 기업별로 한달씩 끊어서 캔들스틱 차트 그리기 

no_img = []

for j in tqdm(range(len(excel_list))):
  df = pd.read_excel(path + '/' + excel_list[j])
  print(excel_list[j])

  for i in tqdm(range(len(start_list))):
    start = start_list[i]
    end = start + relativedelta(months=1) - timedelta(days=1)

    df2 = df.loc[(df['Date']>=start)& (df['Date']<=end)]

    #차트 레이아웃 설정
    fig = plt.figure(figsize=(8, 5))   #최초의 창 크기를 10X10으로 설정, 크기를 설정하지 않으면 아무것도 안나옴
    #ax_main 실제로 데이터가 그려지는 영역
    ax_main = fig.add_subplot(1,1,1)   #1,1,1의 의미는 전체 창을 1X1로 쪼개고, ax_maindmf 생성함


    if (df2.empty == False) :

      if (df2['Open']==0).any() ==False :
        # 캔들차트 그리기
        candlestick2_ohlc(ax_main, df2['Open'], df2['High'], df2['Low'], df2['Close'], width=0.5, colorup='r', colordown='b')

        # x축 없애기
        ax_main.axes.get_xaxis().set_visible(False)
        # y축 없애기
        ax_main.axes.get_yaxis().set_visible(False)

        plt.savefig(path + '_img'+ '/' + excel_list[j].split('.')[0] + '_' + str(start_list[i].month)+'.jpg', dpi=300, bbox_inches='tight')

        plt.close()   #plt.close(fig)
        time.sleep(3)

      else :
        no_img.append(excel_list[j].split('.')[0] + '_' + str(start_list[i].month))
        pass

    else : 
      no_img.append(excel_list[j].split('.')[0] + '_' + str(start_list[i].month))
      pass
```
