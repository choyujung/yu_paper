```python
def resave_imagename(df):
  new_name =[]
  company = []
  for i in range(len(df)):
    com_name = unicodedata.normalize('NFC', df['Company'][i])
    name  = unicodedata.normalize('NFC', df['Image_name'][i])
    new_name.append(name)
    company.append(com_name)

  df['Image_name'] = new_name
  df['Company'] = company
```
